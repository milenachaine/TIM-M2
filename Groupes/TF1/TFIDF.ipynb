```python
import xml.etree.ElementTree as ET
import string
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
import numpy as np
import pandas as pd
```

```python

#rÃ©cuperer le texte
fic='/Users/nino/Desktop/cour_M2/StatistiqueCorpus/TIM-M2/Corpus/all123.xml'
f=open(fic,'r')
xmlcorpus = ET.parse(f)
docs = xmlcorpus.getroot().getchildren()
data=[]
w=open('/Users/nino/Desktop/xmltocsv.csv', 'w') 
for doc in xmlcorpus.getroot():
    if doc.get('class') == 'fake':
        texte="fake"+"\t"+doc.find('text').text
        w.write(texte)
        data.append(texte)
    if doc.get('class') == 'trusted':
        texte="trusted"+"\t"+doc.find('text').text
        w.write(texte)
        data.append(texte)
```

```python
# split sur chaque texte
data_split = [x.split('\t') for x in data if len(x)>=1]

```

```python
texts = [x[1] for x in data_split]
target = [x[0] for x in data_split]

# remplacer fake par 1 et trusted en 0
target = [1. if x=='fake' else 0. for x in target]
```

```python
# Normalize text
# Lower case
texts = [x.lower() for x in texts]
# supp la ponctuation
texts = [''.join(c for c in x if c not in string.punctuation) for x in texts]
# supp les nombre
texts = [''.join(c for c in x if c not in '0123456789') for x in texts]
# enlever les espaces en plus
texts = [' '.join(x.split()) for x in texts]
len(texts)
```

```python
# tokeniez le texte avec nltk.word_tokenize
def tokenizer(text):
    words = nltk.word_tokenize(text)
    return words
```

```python
# pour voir le nombre de words 
string=str(texts)
len(tokenizer(string))[10590]
```

```python
#TF-IDF en sklearn
tfidf = TfidfVectorizer(tokenizer=tokenizer, stop_words='english', max_features=10939)
sparse_tfidf_texts = tfidf.fit_transform(texts)
#len of sparse matrix
sparse_tfidf_texts.shape
```
